```python
b1 = (12 * 0.2)**0.5 + 0.5
b1
```

```python
b2 = -(12 * 0.2)**0.5 + 0.5
b2
```

Так как $b$ - правая граница, то $b  \thickapprox 2.05$

$M(B) = \frac{a + b}{2}$

```python
a = 0.5
b = b1
M = (a + b) / 2
M
```

5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

$z = \frac{x-M(X)}{\sigma} = \frac{x-M(X)}{\sqrt{D(X)}}$

```python
x = 190
z = (x - 178) / 25**0.5
z
```

Таким образом, рост человека равный 190 см отклоняется от математического ожидания на $2.4 \sigma$.
